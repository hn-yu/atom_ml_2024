# 项目二：机器学习模型预测材料性质

机器学习是人工智能领域发展最快的一个分支，涉及计算机、数学模型和统计学等多个领域。机器学习能够通过拟合历史数据，建立数学模型，从数据中有效地学习知识，为我们的决策提供见解。机器学习已经广泛应用于各个领域，包括人脸识别、自然语言处理、高效的网页信息检索、机器人和无人驾驶。目前，机器学习也已经在材料科学、物理、化学 等领域中开始流行起来。传统上机器学习可分为两大类，即监督学习和无监督学习。监督学习通常用于回归和分类两类模型，即对历史数据集中样本的输入、输出建立特定的数学模型；然后通过优化算法求解模型本身的参数，寻找模型参数的这个过程称为“学习”；最后再根据学习到的数学模型对新的输入数据预测对应的输出。

- 回归模型是一种用于预测数值型连续输出的模型，例如根据石墨烯的结构预测石墨烯的带隙大小。常见的回归模型包括线性回归、LASSO模型、岭回归、支持向量回归、核岭回归、集成回归模型。在分类的任务中，预测结果是一个类别，其可能性是有限的，例如根据MoS2的晶体结构预测其是否是半导体，常见的分类模型有逻辑回归、支持向量机、随机森林和神经网络等。

- 在无监督学习中，数据集中的每个样本没有对应的输出，只有输入。无监督学习通常用于分析数据本身的规律， 主要包括聚类和降维。聚类用于将相似的数据自动归为相同的类；降维则用于去除数据的冗余，保留数据中重要的信息，常见的算法包括 Kmeans聚类，主成分分析, 奇异值分解。

### 回归模型性能评价指标

- 均方误差（Mean Squared Error, MSE）：衡量预测值与实际值之间差异的常用指标，定义为预测误差的平方和的平均值。MSE能够反映预测值与实际值之间的平均误差大小。
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

- 均方根误差（Root Mean Squared Error, RMSE）：MSE的平方根，提供了与原始数据相同单位的误差度量。RMSE能够量化模型预测的波动性。
$$RMSE = \sqrt{MSE}$$

- 平均绝对误差（Mean Absolute Error, MAE）：所有单个观测值与算术平均值的偏差的绝对值的平均。MAE能够比较准确反映实际预测误差的大小。
$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

- 决定系数（Coefficient of Determination, $R^{2}$ 或 R Squared）：评估回归模型拟合优度的统计指标，取值范围在0到1之间。当$R^{2}=1$时，模型的预测值与实际观测值完全一致；越接近于1，模型拟合的效果越好。 MSE 越小，$R^{2}$ 越大。
$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_{i})^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}=1 - \frac{MSE}{\frac{1}{n} \sum_{i=1}^{n} (y_i - \bar{y})^2}$$

- 调整后的决定系数（Adjusted R-Squared）：调整后的$R^{2}$是对$R^{2}$ 的一种修正，考虑了模型中自变量的数量。调整后的$R^{2}$可以避免由于模型中自变量数量的增加而导致$R^{2}$值虚高的问题。
$$\text{Adjusted } R^2 = 1 - (1 - R^2) \frac{n-1}{n-p-1}$$


其中$n$、$y_{i}$ 、$\hat{y}_{i}$、$\bar{y}$和$p$分别表示样本数量、第$i$个样本的实际标签、第$i$个样本的模型预测值、所有样本标签的均值和模型中使用的特征数量。

### 分类模型性能评价指标


- 混淆矩阵（Confusion Matrix）：混淆矩阵是一种可视化工具，用于详细展示模型的分类结果。它包括真正例、假正例、真负例和假负例的数量，有助于分析模型在不同类别上的表现。

| 实际/预测 | 正例 (Postive prediction)  | 负例 (Negative prediction)|
|-----------|----------------------|----------------------|
| 正例 (Positive label)| 真正例 (True Positive, TP) | 假负例 (False Negative, FN)|
| 负例 (Negative label)| 假正例 (False Positive, FP)| 真负例 (True Negative, TN) |

- 准确率（Accuracy）：表示正确分类的样本占总样本数的比例。它是最直观、最常用的评价指标之一。准确率越高，模型的分类性能越好。但是，当数据集不平衡时，准确率可能不是一个好的评价指标。
$$Accuracy = (TP + TN) / (TP + TN + FP + FN)$$

- 精确率（Precision）：表示被正确预测为正例的样本占所有预测为正例的样本的比例。它衡量了模型预测正例的准确性。精确率越高，模型预测正例时误报的情况就越少。
$$Precision = TP / (TP + FP)$$

- 召回率（Recall）：召回率表示被正确预测为正例的样本占所有实际正例样本的比例。它衡量了模型找出真正正例的能力。召回率越高，模型找出正例的能力就越强。
$$Recall = TP / (TP + FN)$$

- F1分数：F1分数是精确率和召回率的调和平均值，用于综合评价模型的性能。F1值越高，表示模型在精确率和召回率上达到了一个较好的平衡。
$$F1 = 2 * (Precision * Recall) / (Precision + Recall)$$

- ROC曲线（Receiver Operating Characteristic Curve）：通过绘制真正例率（TPR）与假正例率（FPR）之间的关系来展示模型的性能。

- AUC值（Area Under the Curve）：是ROC曲线下的面积，用于量化模型的整体性能。AUC值越大，表示模型的分类性能越好。AUC值不受数据集不平衡的影响，因此在某些情况下比准确率更适用。

在这一部分，我们将学习如何使用matminer从pymatgen对象中构建机器学习描述符。

Featurizers（特征生成器）的一般概念是，它们接受材料Pymatgen对象（例如，pymatgen Composition）并输出一个向量。

$$f(Fe_2O_3) -> [1.5, 7.8, 9.1,0.09]$$

- 成分（Composition）
- 晶体结构（Crystal structure）
- 晶体位点（Crystal sites）
- 带结构（Bandstructure）
- 态密度（Density of states）

根据特征生成器的不同，返回的特征可能是数值、分类或混合向量、矩阵或其他pymatgen对象。Matminer拥有超过60个特征生成器，其中大多数是根据发表在同行评议论文中的方法实现的。您可以在Matminer网站上找到完整的[特征生成器列表](https://hackingmaterials.lbl.gov/matminer/featurizer_summary.html)。在本教程中，我们将介绍常用的特征生成器方法。

### 核岭回归模型

#### 最小二乘法
最小二乘法是一种线性回归方法，用线性模型拟合数据点$x$和预测连续变量$y$，其基本思想是通过最小化实际观测值与模型预测值之间的残差平方和来找到最佳拟合直线（或曲线）。在最小二乘法中，我们通过调整模型参数（斜率和截距）来使残差平方和最小化。

具体来说，最小二乘法的步骤如下：
1. 定义线性模型：假设线性关系 y = mx + b，其中 y 是因变量，x 是自变量，m 是斜率，b 是截距。
2. 计算预测值：根据给定的自变量 x 和模型参数 m、b 计算预测值 y_pred。
3. 计算残差：计算每个观测值的残差，即实际观测值与预测值之间的差异。
4. 最小化残差平方和：通过最小化残差的平方和来确定最佳的模型参数 m 和 b。
5. 求解参数：通过数学方法（如求导）来找到使残差平方和最小化的最佳参数值。

最小二乘法在矩阵形式下表示为：

$$
\begin{align*}
\text{minimize} \quad & \Vert \mathbf{y} - \mathbf{X}\boldsymbol{\beta} \Vert_2^2 \\
\text{subject to} \quad & \boldsymbol{\beta} \in \mathbb{R}^p
\end{align*}
$$

其中，
- $\mathbf{y}$ 是一个 $n \times 1$ 的矩阵，由数据集中每个样本的标签$y$组成；
- $\mathbf{X}$ 是一个 $n \times p$ 的矩阵，矩阵的每一行$X_{i}$包括一个样本的全部特征：$x_1, x_2,…,x_p$；
- $\boldsymbol{\beta}$ 是一个 $p \times 1$ 的矩阵，是模型参数；
- $\Vert \cdot \Vert_2$ 表示向量的2范数，即欧几里得范数。

最小二乘法的解在矩阵形式下表示为：
$$
\boldsymbol{\beta} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}
$$

当$\mathbf{X}$不是列满秩时，或者某些列之间的线性相关性比较大时，$\mathbf{X}^T\mathbf{X}$ 的行列式接近于0，即$\mathbf{X}^T\mathbf{X}$ 接近于奇异，上述问题变为一个不适定问题，此时，计算$(\mathbf{X}^T\mathbf{X})^{-1}$ 时误差会很大，**传统的最小二乘法缺乏稳定性与可靠性。**


#### 岭回归

岭回归通过在损失函数中加入一个L2正则化项来解决这个问题：

$$
\begin{align*}
\text{minimize} \quad & \Vert \mathbf{y} - \mathbf{X}\boldsymbol{\beta} \Vert_2^2 + \lambda \Vert \boldsymbol{\beta} \Vert_2^2 \\
\text{subject to} \quad & \boldsymbol{\beta} \in \mathbb{R}^p
\end{align*}
$$

岭回归的解在矩阵形式下表示为，其中$\mathbf{I}$为单位矩阵：
$$
\boldsymbol{\beta} = (\mathbf{X}^T\mathbf{X} + \lambda \mathbf{I})^{-1}\mathbf{X}^T\mathbf{y} = \mathbf{X}^T(\lambda \mathbf{I} + \mathbf{X}\mathbf{X}^{T})^{-1}\mathbf{y}
$$

岭回归通过引入L2正则化项，使[参数的值限制到很小的范围](https://mp.weixin.qq.com/s?__biz=MzU0MDQ1NjAzNg==&mid=2247541375&idx=1&sn=2c4e882d69e0cbec156eec71a38cf599&chksm=fb3a8774cc4d0e62899bb78208be93d1bd2fe3798090c00c9a9ae078a2cd3bd98a6444594cb4&scene=27)，但不为0，提高了模型预测的稳定性，同时也能很好解决多重共线性问题。在实际的使用过程中，需要适当地调整正则化参数使模型达到较好的预测效果。

#### 核岭回归
核方法的原理：
核方法的核心思想是将数据映射到高维空间，使得在原始空间中非线性可分的数据在新空间中线性可分。核方法的精妙之处在于，它不需要显式地计算出映射后的数据特征，而是通过一个核函数直接计算数据特征在高维空间中的内积。
$$
\boldsymbol{\beta} = \mathbf{X}^T(\lambda \mathbf{I} + \mathbf{X}\mathbf{X}^{T})^{-1}\mathbf{y} = \mathbf{X}^T(K + \lambda \mathbf{I})^{-1}\mathbf{y}
$$
$$
\alpha := (K + \lambda \mathbf{I})^{-1}\mathbf{y}
$$
$$
\boldsymbol{\beta} =  \mathbf{X}^T \alpha = \sum_i^{n}\alpha_{i}X_i
$$

对于一个新样本$X^{*}$，预测值为新样本与训练样本的核函数线性组合：
$$
y^{*} = \boldsymbol{\beta}^{T}X^{*} = \sum_i^{n}\alpha_{i}X_i^{T}X^{*} = \sum_i^{n}\alpha_{i}k(X^{*}, X_{i})
$$

##### 常用核函数：
- 径向核函数(rbf)
$$K(x, y) = exp(-gamma × ||x - y||^2)$$
- 多项式核函数(poly)
$$k(x, xi) = ((x · xi) + R)^d$$
- 拉普拉斯核函数(laplacian)
$$K(x, y) = exp(-gamma × |x - y|)$$

#### 上述例子中，特征归一化处理有助于提高模型的预测精度。

在机器学习中，数据归一化是一种常见的数据预处理技术，主要目的是将不同尺度或范围的数值转换到一个共同的范围上。通常能达到如下效果：

1、**加快收敛速度**：在梯度下降等优化算法中，如果特征的尺度差异很大，可能会导致优化过程缓慢或不收敛。归一化可以帮助算法更快地收敛。

2、**提高模型性能**：某些机器学习算法（如SVM、KNN）对特征的尺度非常敏感。如果特征的范围差异很大，可能会影响这些算法的性能。归一化可以提高这些算法的准确性和鲁棒性。

3、**正态化假设**：一些机器学习算法（如线性回归）假设数据是正态分布的。归一化有助于使数据更接近正态分布，尽管这不是所有情况下的必要条件。

4、**计算稳定性**：在数值计算中，极端值或大范围的数据可能会导致数值不稳定。归一化可以减少这种风险，提高计算的稳定性。

## 特征筛选
### 移除高度线性相关的特征中皮尔逊相关系数大于其平均皮尔逊相关系数的特征

````python
"""
函数的主要步骤如下：

1、初始化一个列表features_to_keep，包含相关性矩阵的所有特征。

2、对于相关性矩阵中的每一列（即每一个特征），执行以下操作：

    - 获取与该特征高度相关的其他特征，即皮尔森相关系数大于阈值的特征。
    - 如果存在高度相关的特征，计算这些特征的平均相关性。
    - 对于每一个高度相关的特征，如果其相关性大于平均相关性，则将其添加到待移除特征列表features_to_remove中。
    - 更新features_to_keep列表，移除待移除特征列表中的特征。

3、返回features_to_keep列表，即保留的非冗余特征。

"""

def remove_redundant_features_1(corr_matrix, threshold=0.9):
    features_to_keep = list(corr_matrix.columns)
    
    for col in corr_matrix.columns:
        corr_values = corr_matrix.loc[col]
        highly_correlated = corr_values[corr_values >  threshold].index # 获取皮尔森相关系数大于threshold的特征的特征名，即高度相关的特征
        
        if len(highly_correlated) > 1:
            avg_corr = corr_values[highly_correlated].mean() 
            features_to_remove = []
            
            for feature in highly_correlated:
                if corr_values[feature] > avg_corr:
                    features_to_remove.append(feature)
            
            features_to_keep = [f for f in features_to_keep if f not in features_to_remove]
    
    return features_to_keep
````

### 保留高度线性相关特征中方差最大的特征

````python
"""
根据皮尔逊相关系数的大小移除dataframe中高度相关的特征（例如皮尔逊相关系数的绝对值大于0.9），高度相关的特征中仅保留一个，保留方差最大的那个特征。
"""

def remove_redundant_features_2(df, threshold=0.9):
    
    # 计算皮尔逊相关系数矩阵
    corr_matrix = df.corr().abs()

    # 初始化一个集合来存储要删除的特征
    features_to_remove = set()

    # 遍历相关系数矩阵的上三角部分（不包括对角线）
    for i in range(len(corr_matrix.columns)):
        for j in range(i + 1, len(corr_matrix.columns)):
            if corr_matrix.iloc[i, j] > threshold:
                # 如果两个特征的相关性大于阈值，保留方差较大的特征
                if df[corr_matrix.columns[i]].var() < df[corr_matrix.columns[j]].var():
                    features_to_remove.add(corr_matrix.columns[i])
                else:
                    features_to_remove.add(corr_matrix.columns[j])

    return df.drop(features_to_remove, axis=1)
````
