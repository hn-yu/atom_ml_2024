## 项目一：基于Materials Project材料数据，构建自定义数据集

### [Material Project简介](https://next-gen.materialsproject.org/materials)

Materials Project是由Lawrence Berkeley国家实验室发起的开放性材料学数据库和计算平台。该项目旨在为研究人员和工业界提供免费、开放、高质量的材料数据和计算工具，以加速材料科学的发展。其主要特点包括：
- 全面的材料数据：Materials Project包含超过1亿个化合物的材料数据，涵盖了元素、晶体结构、能带结构、密度、形变、热力学等方面的信息。
- 开放的计算平台：Materials Project提供基于密度泛函理论（DFT）的计算工具，帮助研究人员预测新材料的性质和特性。
- 数据可视化和搜索功能：Materials Project提供直观的数据可视化和搜索工具，帮助用户查找和理解材料数据。

通过Materials Project，研究人员可以快速获取高质量的材料数据和计算工具，加速材料科学的研究和应用。该项目已成为材料科学领域的重要资源，吸引了众多科学家和工业界人士的关注和参与。

### 大规模数据下载

当从Materials Project API下载大规模数据集时，需要注意API的速率限制，初始速率为每秒25次。以下是一些下载大型数据集时的提示，以确保快速下载而不触发限制：

- 1、尽可能将查询参数值作为列表传递，以获取多个材料的数据。避免使用循环遍历材料ID值，并为每个单独的材料使用get_data_by_id方法。相反，将材料ID作为列表传递给search方法，一次性获取多个材料的数据。

- 2、在请求数据之前，使用has_props关键字可以查找具有所需属性数据的材料。这样做可以避免在请求不存在或不包含所需属性的材料数据时浪费查询。因此，你应该先确定哪些材料具有你所寻找的数据。以下是一个示例查询，用于获取具有介电常数和态密度数据的所有材料的材料ID值：

```python
with MPRester("yDcEWkj9rYpuqzmlwoNUDCaiTbn3sqLe") as mpr:
    # 可使用??HasProps查询所有属性值
    from emmet.core.summary import HasProps
    docs = mpr.summary.search(has_props=[HasProps.dielectric, HasProps.dos], 
                              fields=["material_id"])
```

- 3、为了减轻服务器负载并加快数据检索速度，建议尽可能将返回的数据限制在感兴趣的特定字段上。例如，如果你只对材料ID、体积和元素列表感兴趣，可以将这些值传递给fields参数：

针对大型、长时间运行或频繁重复的查询，请尽可能在本地创建数据副本，并仅使用API一次性检索数据。这样可以提高你自己的分析速度，并避免对Materials Project服务器造成不必要的负载。

### 案例
获取感兴趣的材料性质，并将结构和对应的性质保存在本地

```python
from mp_api.client import MPRester
import pickle

with MPRester("yDcEWkj9rYpuqzmlwoNUDCaiTbn3sqLe") as mpr:
    docs = mpr.summary.search(
                              fields=["energy_above_hull", "density", "volume", "band_gap", "efermi", "structure", "material_id"]
    )
    
strucs = dict()

for mp in docs:
    material_id = mp.material_id
    
    strucs[material_id] = dict([
        ("energy_above_hull", mp.energy_above_hull),
        ("density", mp.density),
        ("volume", mp.volume),
        ("band_gap", mp.band_gap),
        ("efermi", mp.efermi),
        ("structure", mp.structure)
    ])
    
    # 保存所有结构到本地./mp_cifs_20240627目录（注意需要在当前路径下先创建mp_cifs_20240627目录）
    mp.structure.to('./mp_cifs_20240627/'+material_id+'.cif')
    

with open('data.pkl', 'wb') as file:
    # 将strucs数据保存到本地
    pickle.dump(strucs, file)
```
