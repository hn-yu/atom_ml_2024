# Matplotlib 快速入门教程

## [Matplotlib 简介](https://matplotlib.org/stable/users/explain/quick_start.html#axis)
Matplotlib是一个用于Python编程语言的绘图库，它提供了创建静态、交互式和动画可视化图形的功能。Matplotlib的主要模块是`pyplot`，它提供了类似于MATLAB的绘图接口，使得创建复杂的图形变得简单直观。

### 主要特点
&ensp;&ensp;1、多样化的图形类型：Matplotlib支持各种图形类型，包括折线图、散点图、柱状图、直方图、极坐标图、3D图等。

&ensp;&ensp;2、高度可定制：Matplotlib提供了丰富的定制选项，允许用户调整图形的各个方面，如线条样式、颜色、标记、轴标签、标题、图例等。

&ensp;&ensp;3、交互式绘图：Matplotlib支持交互式绘图，允许用户在图形上进行缩放、平移等操作，以便更好地探索数据。

&ensp;&ensp;4、与NumPy和Pandas集成：Matplotlib可以与NumPy和Pandas等流行的Python数据处理库无缝集成，方便用户直接使用这些库的数据进行绘图。

&ensp;&ensp;5、扩展性：除了内置功能外，Matplotlib还支持第三方插件，可以扩展其功能以适应特定需求。

### Matplotlib的基本概念：Figure、Axes和Axis

&ensp;&ensp;1、**Figure（图形）**：Figure是整个图形的容器，它负责管理所有的子组件，如轴、文本、线条等。一个Figure可以包含一个或多个Axes对象。在Matplotlib中，我们通常通过`plt.subplots()`创建一个Figure对象。

&ensp;&ensp;2、**Axes（轴）**：它表示图形中的一个或多个轴区域。Axes 对象负责管理绘制数据、坐标轴以及其他图形元素。一个 Figure 可以包含多个 Axes 对象，每个 Axes 对象可以包含两个（或三个，对于3D图形）Axis 对象。在设置图形属性和绘制数据时，我们通常会与 Axes 对象进行交互。

&ensp;&ensp;3、**Axis（坐标轴）**：Axis对象表示Axes中的坐标轴，负责设置刻度、刻度标签和坐标轴范围。一个Axes对象包含两个（或三个，对于3D图形）Axis对象，分别表示x轴、y轴（和z轴）。我们可以通过Axes对象的`set_xlabel()`、`set_ylabel()`等方法设置坐标轴标签。

总结一下，Figure是整个图形的容器，负责管理所有子组件；Axes是用于绘制数据和坐标轴的矩形区域；Axis表示Axes中的坐标轴，负责设置刻度和标签。这些概念在创建和自定义Matplotlib图形时非常重要。

## 2. 引入库
在Python脚本中，我们需要引入必要的库：

````python
import matplotlib.pyplot as plt
import numpy as np
````

## 3. 创建一个简单的图形
要创建一个简单的折线图，我们可以使用`pyplot.subplots()`方法创建一个图和轴，然后使用Axes.plot()方法绘制数据：

````python
# 创建一个包含单个轴（axes）的图
fig, ax = plt.subplots()

# 在轴上绘制一些数据。np.linspace用于在指定的区间内生成等差数列
x = np.linspace(-6, 6, 500)
y = np.sin(5*x)
ax.plot(x, y)

# 显示图形
plt.show()
````

## 6. 子图

- 1、`plt.subplots()`允许你在创建图形时一次性添加多个子图，并方便地设置共享轴和其他属性。它返回一个包含子图的数组，使得操作多个子图更加简便。
- 2、`fig.add_subplot()`允许你在已创建的图形中逐个添加子图。你可以更灵活地控制子图的布局和属性，但需要手动设置共享轴和其他参数。

````python
# 创建示例数据集
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建共享x轴的子图
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

# 在第一个子图中绘制y1
ax1.plot(x, y1)
ax1.set_ylabel('$\sin(x)$')

# 在第二个子图中绘制y2
ax2.plot(x, y2)
ax2.set_ylabel('$\cos(x)$')
ax2.set_xlabel('x')

plt.show()
````

## 7、绘图练习

- **绘制LJ势函数：**
$$V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]$$

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator

def lj_potential(r, epsilon=1.0, sigma=1.0):
    return 4 * epsilon * ((sigma / r) ** 12 - (sigma / r) ** 6)

# 设置r的范围和步长
r_min = 0.5
r_max = 3
r_step = 0.01
r = np.arange(r_min, r_max, r_step)

# 计算LJ势能
v = lj_potential(r)

# 创建一个图形和一个子图（Axes对象）
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制LJ势能曲线
ax.plot(r, v, color='blue', linewidth=2, linestyle='-', label='LJ Potential')

# 设置坐标轴标签和刻度标签的字体和字号
ax.set_xlabel('Distance (r)', fontsize=14, fontname='Arial')
ax.set_ylabel('LJ Potential (V(r))', fontsize=14, fontname='Arial')

# 设置x轴和y轴刻度位置
xticks = ax.get_xticks()  # 获取当前x轴的主要刻度位置，并将它们存储在xticks变量中。
yticks = ax.get_yticks()
ax.xaxis.set_major_locator(FixedLocator(xticks))  # 使用FixedLocator设置x轴的主要刻度位置。设置x轴的刻度位置。
ax.yaxis.set_major_locator(FixedLocator(yticks))

# 设置x轴和y轴刻度标签的字体、字号和方向
ax.set_xticklabels(xticks, fontsize=12, fontname='Arial', rotation=45)
ax.set_yticklabels(yticks, fontsize=12, fontname='Arial', rotation=45)

# 设置标题
ax.set_title('Lennard-Jones Potential', fontsize=16, fontname='Arial')

# 添加网格线
ax.grid(True, linestyle='--', alpha=0.7)

# 添加图例
ax.legend(loc='best', fontsize=12)

# 显示图形
plt.show()
````
