# Matplotlib 快速入门教程

## Matplotlib 简介
Matplotlib是一个用于Python编程语言的绘图库，它提供了创建静态、交互式和动画可视化图形的功能。Matplotlib的主要模块是`pyplot`，它提供了类似于MATLAB的绘图接口，使得创建复杂的图形变得简单直观。

### 主要特点
&ensp;&ensp;1、多样化的图形类型：Matplotlib支持各种图形类型，包括折线图、散点图、柱状图、直方图、极坐标图、3D图等。

&ensp;&ensp;2、高度可定制：Matplotlib提供了丰富的定制选项，允许用户调整图形的各个方面，如线条样式、颜色、标记、轴标签、标题、图例等。

&ensp;&ensp;3、交互式绘图：Matplotlib支持交互式绘图，允许用户在图形上进行缩放、平移等操作，以便更好地探索数据。

&ensp;&ensp;4、与NumPy和Pandas集成：Matplotlib可以与NumPy和Pandas等流行的Python数据处理库无缝集成，方便用户直接使用这些库的数据进行绘图。

&ensp;&ensp;5、扩展性：除了内置功能外，Matplotlib还支持第三方插件，可以扩展其功能以适应特定需求。

### Matplotlib的基本概念：Figure、Axes和Axis

&ensp;&ensp;1、**Figure（图形）**：Figure是整个图形的容器，它负责管理所有的子组件，如轴、文本、线条等。一个Figure可以包含一个或多个Axes对象。在Matplotlib中，我们通常通过`plt.subplots()`创建一个Figure对象。

&ensp;&ensp;2、**Axes（轴）**：它表示图形中的一个或多个轴区域。Axes 对象负责管理绘制数据、坐标轴以及其他图形元素。一个 Figure 可以包含多个 Axes 对象，每个 Axes 对象可以包含两个（或三个，对于3D图形）Axis 对象。在设置图形属性和绘制数据时，我们通常会与 Axes 对象进行交互。

&ensp;&ensp;3、**Axis（坐标轴）**：Axis对象表示Axes中的坐标轴，负责设置刻度、刻度标签和坐标轴范围。一个Axes对象包含两个（或三个，对于3D图形）Axis对象，分别表示x轴、y轴（和z轴）。我们可以通过Axes对象的`set_xlabel()`、`set_ylabel()`等方法设置坐标轴标签。

总结一下，Figure是整个图形的容器，负责管理所有子组件；Axes是用于绘制数据和坐标轴的矩形区域；Axis表示Axes中的坐标轴，负责设置刻度和标签。这些概念在创建和自定义Matplotlib图形时非常重要。

## 2. 引入库
在Python脚本中，我们需要引入必要的库：

````python
import matplotlib.pyplot as plt
import numpy as np
````

## 3. 创建一个简单的图形
要创建一个简单的折线图，我们可以使用`pyplot.subplots()`方法创建一个图和轴，然后使用Axes.plot()方法绘制数据：

````python
# 创建一个包含单个坐标轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
ax.plot(x, y)

# 显示图形。
plt.show()
````

````python
# 创建一个包含单个坐标轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = np.linspace(-6, 6, 500)
y = np.sin(5*x)
ax.plot(x, y)

# 显示图形。
plt.show()
````

## 4. 自定义图形
Matplotlib提供了许多自定义选项，例如更改线条样式、添加图例、设置轴标签等。以下是一些常见的自定义方法：

<span style="font-size: 1.3em">&bull; 更改线条样式：</span>

`ax.plot(x, y, linestyle='--', linewidth=2, marker='o')`

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
ax.plot(x, y, linestyle='--', linewidth=2, marker='o')

# 显示图形。
plt.show()
````

<span style="font-size: 1.3em">&bull; 添加标题和轴标签：</span>

```
ax.set_title('My Plot')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
ax.plot(x, y, linestyle='--', linewidth=2, marker='o')

# 添加标题和轴标签
ax.set_title('My Plot')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# 显示图形。
plt.show()
````

<span style="font-size: 1.3em">&bull; 添加图例：</span>

```
ax.plot(x, y, label='My Data')
ax.legend()
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
ax.plot(x, y, linestyle='--', linewidth=2, marker='o', label='My Data')

# 添加标题和轴标签
ax.set_title('My Plot')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# 添加图例
ax.legend()

# 显示图形。
plt.show()
````

<span style="font-size: 1.3em">&bull; 设置轴范围和刻度：</span>

```
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_yticks([1, 2, 3, 4, 5])
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
ax.plot(x, y, linestyle='--', linewidth=2, marker='o', label='My Data')

# 添加标题和轴标签
ax.set_title('My Plot')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# 添加图例
ax.legend()

# 设置轴范围
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_yticks([0, 1, 2, 3, 4, 5])

# 显示图形。
plt.show()
````

## 5. 其他图形类型
Matplotlib支持多种图形类型，例如散点图、柱状图、直方图等。以下是一些示例：

<span style="font-size: 1.3em">&bull; 散点图：</span>

```
ax.scatter(x, y)
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

# 散点图
ax.scatter(x, y, label='My Data')

# 添加标题和轴标签
ax.set_title('My Plot')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# 添加图例
ax.legend()

# 设置轴范围
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_yticks([1, 2, 3, 4, 5])

# 显示图形。
plt.show()
````

<span style="font-size: 1.3em">&bull; 柱状图：</span>

```
ax.bar(x, y)
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
x = ['A', 'B', 'C', 'D']
y = [10, 20, 30, 40]

# 柱状图
ax.bar(x, y)

# 显示图形。
plt.show()
````

<span style="font-size: 1.3em">&bull; 直方图：</span>

```
ax.hist(data, bins=10)
```

````python
# 创建一个包含单个轴的图。
fig, ax = plt.subplots()

# 在轴上绘制一些数据。
data = np.random.randn(100)

# 直方图
ax.hist(data, bins=10)

# 显示图形。
plt.show()
````
