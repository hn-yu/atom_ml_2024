### 1. 自定义ASE的Calculator

接着我们需要一个编写一个ASE的接口来将机器学习力场计算结果传递给ASE，用于进一步的计算。

首先我们要把前两节写过的`AseDataReader`给用起来，这是为了我们能够直接读取ASE的Atoms对象，用来给模型做计算。

直接复制过来就行。

```python
class AseDataReader:
    def __init__(self, cutoff=5.0):            
        self.cutoff = cutoff
        self.neighbor_list = asap3.FullNeighborList(self.cutoff, atoms=None)

    def __call__(self, atoms):
        atoms_data = {
            'num_atoms': torch.tensor([atoms.get_global_number_of_atoms()]),
            'elems': torch.tensor(atoms.numbers),
            'coord': torch.tensor(atoms.positions, dtype=torch.float),
        }
        atoms_data['image_idx'] = torch.zeros((atoms_data['num_atoms'],), dtype=torch.long)
        if atoms.pbc.any():            
            atoms_data['cell'] = torch.tensor(atoms.cell[:], dtype=torch.float)

        pairs, n_diff = self.get_neighborlist(atoms)    
        atoms_data['pairs'] = torch.from_numpy(pairs)
        atoms_data['n_diff'] = torch.from_numpy(n_diff).float()
        atoms_data['num_pairs'] = torch.tensor([pairs.shape[0]])
        
        # 当atoms有能量和力输出时，读取这些输出。这些数据对训练模型是必要的，但是对预测不是必要的。
        try:
            energy = torch.tensor([atoms.get_potential_energy()], dtype=torch.float)
            atoms_data['energy'] = energy
        except (AttributeError, RuntimeError):
            pass
        
        try: 
            forces = torch.tensor(atoms.get_forces(apply_constraint=False), dtype=torch.float)
            atoms_data['forces'] = forces
        except (AttributeError, RuntimeError):
            pass
        
        return atoms_data
    
    def get_neighborlist(self, atoms):        
        self.neighbor_list.check_and_update(atoms)            # 更新结构
        pair_i_idx = []
        pair_j_idx = []
        n_diff = []
        for i in range(len(atoms)):
            indices, diff, _ = self.neighbor_list.get_neighbors(i)
            pair_i_idx += [i] * len(indices)               # local index of pair i
            pair_j_idx.append(indices)   # local index of pair j
            n_diff.append(diff)

        pair_j_idx = np.concatenate(pair_j_idx)
        pairs = np.stack((pair_i_idx, pair_j_idx), axis=1)
        n_diff = np.concatenate(n_diff)
        
        return pairs, n_diff
```

接着我们需要参照ASE中`Calculator`的写法，写一个机器学习力场的`MLCalculator`。

这个`Calculator`中所有输出的性质必须都包含在results这个字典里，接着这些性质便能拿来做一些计算了。

```python
class MLCalculator(Calculator):
    implemented_properties = ["energy", "forces"]

    def __init__(
        self,
        model,
        energy_scale=1.0,
        forces_scale=1.0,
        **kwargs
    ):
        super().__init__(**kwargs)

        self.model = model
        self.model_device = next(model.parameters()).device
        self.cutoff = model.cutoff
        self.ase_data_reader = AseDataReader(self.cutoff)
        self.energy_scale = energy_scale
        self.forces_scale = forces_scale

    def calculate(self, atoms=None, properties=["energy"], system_changes=all_changes):
        # First call original calculator to set atoms attribute
        # (see https://wiki.fysik.dtu.dk/ase/_modules/ase/calculators/calculator.html#Calculator)
        if atoms is not None:
            self.atoms = atoms.copy()       

        model_inputs = self.ase_data_reader(self.atoms)
        model_inputs = {
            k: v.to(self.model_device) for k, v in model_inputs.items()
        }

        model_results = self.model(model_inputs)

        results = {}

        # Convert outputs to calculator format
        results["forces"] = (
            model_results["forces"].detach().cpu().numpy() * self.forces_scale
        )
        results["energy"] = (
            model_results["energy"][0].detach().cpu().numpy().item()
            * self.energy_scale
        )
        self.results = results
```

这个接口写好以后就可以使用ASE来做一些计算了。

### 2. 分子动力学模拟

这里我们定义一个print函数来打印MD过程中的各种状态量。

```python
steps = 0
def printenergy(a=atoms):  # store a reference to atoms in the definition.
    """Function to print the potential, kinetic and total energy."""
    epot = a.get_potential_energy()
    ekin = a.get_kinetic_energy()
    temp = ekin / (1.5 * units.kB) / a.get_global_number_of_atoms()
    global steps
    steps += 1
    status = f"Steps={steps:12.3f} Epot={epot:12.3f} Ekin={ekin:12.3f} temperature={temp:8.2f}"
    print(status)
    with open('MD.log', 'a') as f:
        f.write(status + '\n')
```

```python
from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
from ase.md.langevin import Langevin
from ase.io import Trajectory
import time

# 为MD计算初始化速度分布
MaxwellBoltzmannDistribution(atoms, temperature_K=300)

# 创建MD计算的动力学模块并设置相关参数
dyn = Langevin(atoms, timestep=0.5 * units.fs, temperature_K=300, friction=0.1 / units.fs)
dyn.attach(printenergy, interval=1)

#记录时间
start = time.time()

traj = Trajectory('MD.traj', 'w', atoms)
dyn.attach(traj.write, interval=100)
dyn.run(10000)

print(f"Simulation time: {time.time() - start} s")
```

从以上输出我们可以监测MD的计算，MD计算还将产生一个MD.traj文件，这个文件包含了所有MD产生的结构。可以使用ASE GUI打开查看

```python
from ase.visualize import view
traj = Trajectory('MD.traj')
view(traj[-1], viewer='x3d')
```

什么都看不清楚，这是因为很多水分子运动到box外了。这里我们可以后处理一下，把box外的原子wrap进去，就能更好地查看我们的结构了。

## 项目六：利用不确定性衡量模型预测结果的可靠性

之前我们学习了如何自己亲手搭建一个GNN模型，以及用我们手头的数据来训练并使用这个模型。

这里我们有一个问题，如果用这个模型来计算它没有训练过的体系，它能给出可靠的结果吗？

如果不能的话，我们怎么来衡量模型预测的可靠性呢？

一个最经典的方法就是使用ensemble，也就是多个模型一起预测来获得不确认性的信息。

举个例子，下面的公式表示利用不同模型预测的能量值标准差，来衡量这些模型对一个预测结构的不确定性。

$$
\begin{align*}
    \sigma_E(S) &= \sqrt{\sum^{N_{pred}}_{n=1}(E^{pred} - \bar{E})^2} \\
\end{align*}
$$

同样，我们也可以使用力的标准差（或者叫RMSD）来衡量模型的不确定性，因为力的训练相对能量往往要难很多。

$$
\begin{align*}
    \sigma_F(S) &= \sqrt{\frac{1}{3N_{atoms}N_{pred}}\sum^{N_{pred}}_{n=1}\sum^{N_{atoms}}_{i=1}\sum^{3}_{j=1} ({F}^{pred}_{ij} - \bar{{F}_{ij}})^2}
\end{align*}
$$

那具体怎么操作呢？我们可以单纯地改变`Calculator`的写法，或者是新建一个class `Ensemble`，来处理多个模型的输出，并计算不确定性。这里我们采用后一种方法，前一种方法大家可以自行去试试。

```python
class Ensemble(nn.Module):
    """
    Ensemble model of PaiNN. This is used to get uncertainty information.
    """
    def __init__(self, models: List[nn.Module]) -> None:
        super().__init__()
        self.models = nn.ModuleList([model for model in models])
        self.compute_uncertainty = True if len(models) > 1 else False
        self.cutoff = models[0].cutoff

    def forward(self, input_dict: Dict[str, torch.Tensor], compute_forces: bool=True) -> Dict[str, torch.Tensor]:
        energy = []
        forces = []
        for model in self.models:
            out = model(input_dict, compute_forces)
            energy.append(out['energy'].detach())
            forces.append(out['forces'].detach())
        
        energy = torch.stack(energy)
        forces = torch.stack(forces)
        f_scatter = torch.zeros(input_dict['num_atoms'].shape[0], device=out['energy'].device)
        
        result_dict = {
            'energy': torch.mean(energy, dim=0),
            'forces': torch.mean(forces, dim=0),
        }
        if self.compute_uncertainty:
            result_dict.update({
                'e_var': torch.var(energy, dim=0),
                'e_max': torch.max(energy).unsqueeze(-1),
                'e_min': torch.min(energy).unsqueeze(-1),
                'e_sd': torch.std(energy, dim=0),
                'f_var': f_scatter.index_add(0, input_dict['image_idx'], torch.var(forces, dim=0).mean(dim=1)) / input_dict['num_atoms'],
            })
            result_dict['f_sd'] = result_dict['f_var'].sqrt()
            
        if 'energy' in input_dict.keys():
            e_diff = result_dict['energy'] - input_dict['energy']
            f_diff = result_dict['forces'] - input_dict['forces']
            result_dict.update({
                'e_ae': torch.abs(e_diff),
                'e_se': torch.square(e_diff),
                'f_ae': f_scatter.index_add(0, input_dict['image_idx'], torch.abs(f_diff).mean(dim=1)) / input_dict['num_atoms'],
                'f_se': f_scatter.index_add(0, input_dict['image_idx'], torch.square(f_diff).mean(dim=1)) / input_dict['num_atoms'],
            })

        return result_dict 
```

```python
from pathlib import Path

# load 所有model
models = []
for model_path in Path('models/ensemble').glob('*.pt'):
    compiled = torch.jit.load(model_path)
    models.append(compiled)

# 创建ensemble
ensemble = Ensemble(models)
```

接下来我们便可以创建一个使用ensemble来做计算的calculator。

```python
class MLCalculator(Calculator):
    implemented_properties = ["energy", "forces"]

    def __init__(
        self,
        model,
        energy_scale=1.0,
        forces_scale=1.0,
        **kwargs
    ):
        super().__init__(**kwargs)

        self.model = model
        self.model_device = next(model.parameters()).device
        self.cutoff = model.cutoff
        self.ase_data_reader = AseDataReader(self.cutoff)
        self.energy_scale = energy_scale
        self.forces_scale = forces_scale

    def calculate(self, atoms=None, properties=["energy"], system_changes=all_changes):
        # First call original calculator to set atoms attribute
        # (see https://wiki.fysik.dtu.dk/ase/_modules/ase/calculators/calculator.html#Calculator)
        if atoms is not None:
            self.atoms = atoms.copy()       

        model_inputs = self.ase_data_reader(self.atoms)
        model_inputs = {
            k: v.to(self.model_device) for k, v in model_inputs.items()
        }

        model_results = self.model(model_inputs)

        results = {}

        # Convert outputs to calculator format
        results["forces"] = (
            model_results["forces"].detach().cpu().numpy() * self.forces_scale
        )
        results["energy"] = (
            model_results["energy"][0].detach().cpu().numpy().item()
            * self.energy_scale
        )

        # 新增的uncertainty相关信息
        for k, v in model_results.items():
            if k != "forces" and k != "energy":
                results[k] = v.detach().cpu().item()
        
        self.results = results
```

```python
en_calc = MLCalculator(ensemble)
```

OK。那现在让我们做一个可以实时输出uncertainty的MD模拟。

```python
steps = 0
def print_energy_uncertainty(a=atoms):  # store a reference to atoms in the definition.
    """Function to print the potential, kinetic and total energy."""
    epot = a.get_potential_energy()
    ekin = a.get_kinetic_energy()
    temp = ekin / (1.5 * units.kB) / a.get_global_number_of_atoms()
    results = a.calc.results
    e_var = results['e_var']
    e_sd = results['e_sd']
    f_var = results['f_var']
    f_sd = results['f_sd']
    global steps
    steps += 1
    status = f"Steps={steps:12.3f} Epot={epot:12.3f} Ekin={ekin:12.3f} temperature={temp:8.2f} e_var={e_var:10.3g} e_sd={e_sd:10.3g} f_var={f_var:10.3g} f_sd={f_sd:10.3g}"
    print(status)
    with open('ensemble.log', 'a') as f:
        f.write(status + '\n')

# 为MD计算初始化速度分布
MaxwellBoltzmannDistribution(atoms, temperature_K=300)

# 创建MD计算的动力学模块并设置相关参数
dyn = Langevin(atoms, timestep=0.5 * units.fs, temperature_K=300, friction=0.1 / units.fs)
dyn.attach(print_energy_uncertainty, interval=1)

#记录时间
start = time.time()

traj = Trajectory('ensemble_MD.traj', 'w', atoms)
dyn.attach(traj.write, interval=100)
dyn.run(2000)

print(f"Simulation time: {time.time() - start} s")
```
